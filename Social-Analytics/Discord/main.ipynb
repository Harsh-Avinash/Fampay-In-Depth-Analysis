```python
import discord
import pandas as pd
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
import nest_asyncio

nltk.download('vader_lexicon')
nest_asyncio.apply()

intents = discord.Intents.default()
intents.typing = False
intents.presences = False
intents.messages = True

client = discord.Client(intents=intents)
sia = SentimentIntensityAnalyzer()

# Replace 'your-bot-token' with the token you copied earlier
TOKEN = ''

# Function to perform sentiment analysis
def analyze_sentiment(text):
    sentiment = sia.polarity_scores(text)
    return sentiment

# Connects the bot to the server
@client.event
async def on_ready():
    print(f'{client.user} has connected to Discord!')

    # Replace 'server-id' with your server's ID
    server = client.get_guild('761355612797272094')
    
    # Collect messages and users
    data = []

    for channel in server.text_channels:
        try:
            async for message in channel.history(limit=100):
                sentiment = analyze_sentiment(message.content)
                data.append([message.author.id, message.content, sentiment['compound'], message.created_at])
        except discord.errors.Forbidden:
            continue

    # Create a DataFrame and analyze data
    df = pd.DataFrame(data, columns=['User ID', 'Message', 'Sentiment', 'Timestamp'])

    # Active users
    active_users = df['User ID'].value_counts()

    # Sentiment analysis
    average_sentiment = df['Sentiment'].mean()

    # Print results
    print("Active Users:")
    print(active_users)
    print("\nAverage Sentiment:")
    print(average_sentiment)

client.run(TOKEN)
```
