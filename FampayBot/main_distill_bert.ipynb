```python
from transformers import DistilBertTokenizer, DistilBertForQuestionAnswering
import torch

def chatbot(text, question):
    model_name = "distilbert-base-uncased-distilled-squad"
    tokenizer = DistilBertTokenizer.from_pretrained(model_name)
    model = DistilBertForQuestionAnswering.from_pretrained(model_name)

    encoding = tokenizer.encode_plus(question, text, return_tensors="pt")
    input_ids = encoding["input_ids"]
    attention_mask = encoding["attention_mask"]

    outputs = model(input_ids, attention_mask=attention_mask)
    start_score, end_score = outputs.start_logits, outputs.end_logits

    answer_start = torch.argmax(start_score)
    answer_end = torch.argmax(end_score)

    tokens = tokenizer.convert_ids_to_tokens(input_ids[0])

    answer = ' '.join(tokens[answer_start:answer_end + 1])

    answer = answer.replace(' ##', '').replace('[CLS]', '').replace('[SEP]', '').strip()

    return answer

# Example usage:
text = """Party CapParty FaceWohoo! We just launched India's first ever doodle card for teens.
Arrow Head
FamPay
About
Blog

Download FamApp
banner
spending account
for
teens
adults
everyone

Open Account
10M+ Registered Users
Powered by
TrioVisaRupayUPI
One account for all your spends
Get a card & UPI linked to your FamX account
Payments that are quick, easy & accepted everywhere
Easily track and keep your expenses in check
Keep your bank statements clean & clutter-free

FamCard
Your first and forever account
Make payments without a bank account
Get a customised UPI & card as a teen
Open for everyone ages 11+
upi
UPI payments that go through in seconds
Scan & Pay anywhere with your customised UPI ID
Choose from a range of fun QR themes
Recharge, pay bills and shop online, all within the app

Get your own UPI
rewards
The most rewarding
spending account
Perks of FamX

Cashback & FamCoins on everyday spends

Enjoy exciting discounts & offers from top brands

Win big from spinners and giveaways

upi
FamX Card Title
Numberless card that keeps your info safe
Quick and secure payments with Tap & Pay
Design your own card with doodles
Make ATM withdrawals with your FamX card
Memberships tailor-made for your needs
Hiring
Hiring
Hiring
Hiring
Our support team is
active 24x7
Feel free to chat with our support team whenever you need more clarity

Over a million happy stories
Yeh toh upgrade ka bhi upgrade hai üêê
The new design is so cool and app has become more smooth. The FamApp has everything I need in one place.

Eshaa

We are trusted by
Elevation Capital
Sequoia
Venture Highway
General Catalyst
Y Combinator
Global Founders Capital
Rocketship VC
GreenOaks
Hiring
Want to shape finance
for the next gen?

Join the Fam
Fam
Fam, formerly FamPay, focuses on financial inclusion of the next generation. FamX by Fam, is India's first Spending Account made for everyone. With your own Spending Account, you can now save, spend, earn, and learn everything about money!

¬© 2023 Fam | All rights reserved | Made with üß°

Company
About
Careers
Become a partner
Resources
FAQs
Blog
Press
FamPay-IDFC
Privacy
Terms
Contact
FamX-Trio
Privacy
Terms
FacebookMediumInstagramLinkedInTwitterDiscordSnapchatYoutube
"""

question = "What is famx?"
answer = chatbot(text, question)
print(answer)
```
